#  R Linear Model

## Import statements

```r
library ("ggplot2")
install.packages ("ggplot2")
```

## Drop .csv HERE

```r
dataset = read.csv ("regrex1.csv")
print(dataset)
```

### Starting linear model

```r
model = lm(formula = x ~ y, data = dataset)
```

## Graph

### Scatter Plot

```r
library (ggplot2) 
ggplot() +
    geom_point(aes(x = dataset$ x, y = dataset$ y),color = 'orange')
```

```r
library(tidyverse)
ggplot() +
  geom_point(aes(x = dataset$x, y = dataset$y), color = 'orange') +
  geom_line(aes(x = dataset$y, y = predict(model, newdata = dataset)), color = 'green') +
  ggtitle("x vs y")
```

summary (model)
